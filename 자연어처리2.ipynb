```python
import re

def text_cleaning(text):
    if isinstance(text, str):
        hangul = re.compile('[^ ㄱ-ㅣ 가-힣]+')
        result = hangul.sub('', text)
        return result
    else:
        return ""  # 비어 있는 문자열을 반환하거나 다른 방법을 선택할 수 있습니다.

# 나머지 코드는 그대로 유지
df['ko_text'] = df['text'].apply(lambda x: text_cleaning(x))
del df['text']
df.head()
#특수문자 전부제거 ex)?,!등등
```

```python
from konlpy.tag import Okt

def get_pos(x, custom_tokens):
    tagger = Okt()

    # custom_tokens에 있는 단어나 어구를 치환
    for token in custom_tokens:
        x = x.replace(token, token.replace(" ", "_"))

    pos = tagger.pos(x)
    pos = ['{}/{}'.format(word.replace("_", " "), tag) for word, tag in pos]

    return pos

custom_tokens = ["양이 적어요", "매워요", "가성비", "양이 많고"]  # 원하는 어구를 리스트에 추가
df['pos'] = df['ko_text'].apply(lambda x: get_pos(x, custom_tokens))

print(df['pos'])
```

```python
import csv
import pandas as pd
from collections import Counter

# CSV 파일 경로 설정
csv_file_path = '/content/drive/MyDrive/캡스톤/230920_중식_종로_RV_.csv'  # CSV 파일 경로를 적절히 설정하세요.

# CSV 파일 읽어오기
df = pd.read_csv(csv_file_path, encoding='cp949')

# 문장 패턴 정의
patterns = ["양이 적어요", "매워요", "맵찔", "가성비"]

# 빈도수를 저장할 딕셔너리 초기화
pattern_counts = {pattern: 0 for pattern in patterns}

# 각 텍스트에서 패턴 검색 및 빈도수 계산
for text in df['text']:
    if pd.notna(text):
        for pattern in patterns:
            if pattern in text:
                pattern_counts[pattern] += 1

# 결과 출력
for pattern, count in pattern_counts.items():
    print(f"{pattern}: {count}번 출현")

# 결과를 DataFrame으로 만들어 CSV 파일로 저장
result_df = pd.DataFrame(list(pattern_counts.items()), columns=['패턴', '빈도수'])
result_df.to_csv('pattern_counts.csv', index=False)
```
